# このノートブックについて

日経平均株価のトレンドを分析する。  
前日比でプラスを1とし、マイナスを0としている。  

# ライブラリインポート

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats as stats
import pymc3 as pm
import arviz as az

from bs4 import BeautifulSoup
import urllib.request as req
```

# データスクレイピング、前処理

```python
def get_kabuka_rate(html, table_code, rate_code):
    res = req.urlopen(html)
    soup = BeautifulSoup(res, "html.parser")
    table = soup.select_one(table_code)
    dates = table.find_all(scope="row")
    rates = table.find_all(rate_code)
    
    result_date = []
    for i in dates:
        i = i.string
        result_date.append(i)
        
    result_rate = []
    for i, rate in enumerate(rates):
        if i % 2 == 0:
            result_rate.append(float(rate.string))
    
    dic = {"date" : result_date, "rate" : result_rate}
    return pd.DataFrame(dic)
```

```python
result_df = get_kabuka_rate(html = "https://kabutan.jp/stock/kabuka?code=0000",
                            table_code = "div#stock_kabuka_table",
                            rate_code = "span")
```

```python
result_df.head()
```

```python
# +を1, -を0とする処理
result_bin = []
for i in result_df["rate"]:
    if i >= 0:
        result_bin.append(1)
    else:
        result_bin.append(0)

result_bin[:5]
```

```python
print("平均値", round(np.average(result_bin), 2))
sns.distplot(result_bin, kde=False, bins=3)
```

# ベイズ推定
事前分布は一様のベータ分布とする

```python
with pm.Model() as model:
    theta = pm.Beta("theta", alpha=1, beta=1)
    y = pm.Bernoulli("y", p=theta,observed=result_bin)
    
    trace = pm.sample(1000)
```

```python
az.plot_trace(trace)
```

```python
az.plot_posterior(trace, rope=[0.45, 0.55], ref_val=0.5)
```
